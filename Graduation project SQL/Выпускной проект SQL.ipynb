##  Количество обзоров и средняя оценка каждой книги

```python
books_reviews = '''
WITH book_stats AS (
    SELECT 
           b.book_id,
           b.title,
           COUNT(DISTINCT rev.review_id) AS count_reviews,
           AVG(rat.rating) AS avg_rating
    FROM books b
    LEFT JOIN ratings rat ON b.book_id = rat.book_id
    LEFT JOIN reviews rev ON b.book_id = rev.book_id
    GROUP BY b.book_id
)

SELECT * 
FROM (
    SELECT * 
    FROM book_stats
    ORDER BY avg_rating DESC, count_reviews DESC
    LIMIT 50
) AS top_by_rating

UNION ALL

SELECT * 
FROM (
    SELECT * 
    FROM book_stats
    ORDER BY count_reviews DESC, avg_rating DESC
    LIMIT 50
) AS top_by_reviews
'''


df = showcase(books_reviews)


top_by_rating = df.iloc[:50]
top_by_reviews = df.iloc[50:].reset_index(drop=True)

display(top_by_rating)
display(top_by_reviews)
```

**Вывод:** Максимум обзоров Twilight - 7, однако рейдинг 3.66. Среди книг с высоким рейтингом (5) максималньое количество отзывов - A Dirty Job (Grim Reaper #1) - 4. Две книги с высоким рейтингом имеют 3 обзора. У остальных книг с высоким рейтингом от 0 до 2 обзоров. Среди книг у которых более 4х обзоров максимальный рейтинг у книги Snow Flower and the Secret Fan -  4.636364.

## Издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
publisher_50 = '''
               SELECT pub.publisher AS publisher,
                      COUNT(b.book_id) AS books_cnt
               FROM publishers pub
               LEFT JOIN books b ON pub.publisher_id = b.publisher_id
               WHERE b.num_pages > 50
               GROUP BY publisher
               ORDER BY books_cnt DESC
               LIMIT 5
               '''

showcase(publisher_50)
```

**Вывод:** больше всего книг больше 50 страниц выпустило издательство Penguin Books - 42 книги.

## Автор с самой высокой средней оценкой книг — учитываются только книги с 50 и более оценками

```python
author_rating = '''
WITH book_ratings AS (
     SELECT 
         b.book_id,
         b.author_id,
         AVG(r.rating) AS avg_rating,
         COUNT(r.rating_id) AS rating_count
     FROM books b
     JOIN ratings r ON b.book_id = r.book_id
     GROUP BY b.book_id, b.author_id
     HAVING COUNT(r.rating_id) >= 50
),
author_avg_rating AS (
    SELECT 
        author_id,
        AVG(avg_rating) AS avg_author_rating
    FROM book_ratings
    GROUP BY author_id
)
SELECT 
    a.author,
    ar.avg_author_rating
FROM author_avg_rating ar
JOIN authors a ON ar.author_id = a.author_id
ORDER BY ar.avg_author_rating DESC
LIMIT 3;
'''

showcase(author_rating)
```

**Вывод:** Автор с самым выскоим средним рейтингом: J.K. Rowling/Mary GrandPré. Средний рейтинг: 4.28

## Среднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
avg_reviews = '''
WITH user_reviews AS (
     SELECT r.username,
            COUNT(r.review_id) AS review_count
     FROM reviews r
     JOIN(SELECT username
          FROM ratings
          GROUP BY username
          HAVING COUNT(rating_id)>48
     ) u ON r.username = u.username
     GROUP BY r.username
     )
SELECT AVG(review_count) AS avg_reviews
FROM user_reviews;
'''

showcase(avg_reviews)
```

**Вывод:** среднее количество обзоров среди пользователей, поставивших более 48 оценок - 24.
