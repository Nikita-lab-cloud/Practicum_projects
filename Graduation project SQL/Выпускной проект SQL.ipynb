##  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏

```python
books_reviews = '''
WITH book_stats AS (
    SELECT 
           b.book_id,
           b.title,
           COUNT(DISTINCT rev.review_id) AS count_reviews,
           AVG(rat.rating) AS avg_rating
    FROM books b
    LEFT JOIN ratings rat ON b.book_id = rat.book_id
    LEFT JOIN reviews rev ON b.book_id = rev.book_id
    GROUP BY b.book_id
)

SELECT * 
FROM (
    SELECT * 
    FROM book_stats
    ORDER BY avg_rating DESC, count_reviews DESC
    LIMIT 50
) AS top_by_rating

UNION ALL

SELECT * 
FROM (
    SELECT * 
    FROM book_stats
    ORDER BY count_reviews DESC, avg_rating DESC
    LIMIT 50
) AS top_by_reviews
'''


df = showcase(books_reviews)


top_by_rating = df.iloc[:50]
top_by_reviews = df.iloc[50:].reset_index(drop=True)

display(top_by_rating)
display(top_by_reviews)
```

–ú–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ —á–µ–º –±–æ–ª—å—à–µ –æ–±–∑–æ—Ä–æ–≤ —É –∫–Ω–∏–≥–∏, —Ç–µ–º —Å–ª–æ–∂–Ω–µ–µ –µ–π –ø–æ–ª—É—á–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –æ—Ü–µ–Ω–∫—É, –ø–æ—Ç–æ–º—É —á—Ç–æ –æ–¥–Ω–∞ –Ω–µ–≥–∞—Ç–∏–≤–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –º–æ–∂–µ—Ç —Å–∏–ª—å–Ω–æ –æ–ø—É—Å—Ç–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥. –í–µ—Ä–æ—è—Ç–Ω–æ –æ–±–∑–æ—Ä—ã –≤–ª–∏—è—é—Ç –Ω–∞ –≤–æ—Å–ø—Ä–∏—è—Ç–∏–µ –∏ –æ—Ü–µ–Ω–∫—É –æ–±—ã—á–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. 

## –ò–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü

```python
publisher_50 = '''
               SELECT pub.publisher AS publisher,
                      COUNT(b.book_id) AS books_cnt
               FROM publishers pub
               LEFT JOIN books b ON pub.publisher_id = b.publisher_id
               WHERE b.num_pages > 50
               GROUP BY publisher
               ORDER BY books_cnt DESC
               LIMIT 5
               '''

showcase(publisher_50)
```

**–í—ã–≤–æ–¥:** –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∫–Ω–∏–≥ –±–æ–ª—å—à–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü –≤—ã–ø—É—Å—Ç–∏–ª–æ –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ Penguin Books - 42 –∫–Ω–∏–≥–∏.

## –ê–≤—Ç–æ—Ä —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥ ‚Äî —É—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏

```python
author_rating = '''
WITH book_ratings AS (
     SELECT 
         b.book_id,
         b.author_id,
         AVG(r.rating) AS avg_rating,
         COUNT(r.rating_id) AS rating_count
     FROM books b
     JOIN ratings r ON b.book_id = r.book_id
     GROUP BY b.book_id, b.author_id
     HAVING COUNT(r.rating_id) >= 50
),
author_avg_rating AS (
    SELECT 
        author_id,
        AVG(avg_rating) AS avg_author_rating
    FROM book_ratings
    GROUP BY author_id
)
SELECT 
    a.author,
    ar.avg_author_rating
FROM author_avg_rating ar
JOIN authors a ON ar.author_id = a.author_id
ORDER BY ar.avg_author_rating DESC
LIMIT 3;
'''

showcase(author_rating)
```

**–í—ã–≤–æ–¥:** –ê–≤—Ç–æ—Ä —Å —Å–∞–º—ã–º –≤—ã—Å–∫–æ–∏–º —Å—Ä–µ–¥–Ω–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º: J.K. Rowling/Mary GrandPr√©. –°—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥: 4.28

## –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª—å—à–µ 48 –æ—Ü–µ–Ω–æ–∫

```python
avg_reviews = '''
WITH user_reviews AS (
     SELECT r.username,
            COUNT(r.review_id) AS review_count
     FROM reviews r
     JOIN(SELECT username
          FROM ratings
          GROUP BY username
          HAVING COUNT(rating_id)>48
     ) u ON r.username = u.username
     GROUP BY r.username
     )
SELECT AVG(review_count) AS avg_reviews
FROM user_reviews;
'''

showcase(avg_reviews)
```
